```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
df = pd.read_csv('./data/life-expectancy-years-vs-real-gdp-per-capita-2011us.csv', date_parser='Year', ).set_index('Entity')
df_continents = pd.read_csv('./data/Countries-Continents.csv', index_col='Country')
```

```python
df.update(df_continents)
```

```python
df = df[df['Continent'].notna()].reset_index()
df.head(3)
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=df, x='Year', y='Life expectancy', hue='Continent')
```
