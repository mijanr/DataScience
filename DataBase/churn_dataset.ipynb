```python
import sqlite3
```

```python
conn = sqlite3.connect('data/churn.db')# You can create a new database by changing the name within the quotes
c = conn.cursor()
```

```python
# Create table - CLIENTS
c.execute('''CREATE TABLE CHURN
          (
            RowNumber INT NOT NULL,
            CustomerId INT NOT NULL,
            Surname VARCHAR(20) NULL,
            CreditScore INT NULL,
            Geography VARCHAR(20) NULL,
            Gender VARCHAR(20) NULL,
            Age INT NULL,
            Tenure INT NULL,
            Balance DECIMAL(10,2) NULL,
            NumOfProducts INT NULL,
            HasCrCard INT NULL,
            IsActiveMember INT NULL,
            EstimatedSalary DECIMAL(10,2) NULL,
            Exited INT NULL,
            PRIMARY KEY(RowNumber)
)''')
```

```python
conn.commit()
```
